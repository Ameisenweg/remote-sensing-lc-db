```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# training data
X = np.array([
    [[0.4, 0.3, 0.3], [0.6, 0.5, 0.3], [0.7, 0.4, 0.4]],
    [[0.4, 0.4, 0.3], [0.4, 0.3, 0.3], [0.7, 0.5, 0.4]],
    [[0.0, 0.6, 0.3], [0.0, 0.5, 0.2], [0.6, 0.4, 0.3]]
])

X = np.reshape(X, (9, 3))

# labels
y = np.array([
    [1, 1, 1],
    [0, 1, 1],
    [0, 0, 1]
])

y = y.flatten()  # similar to np.reshape(y, (9,))

# initiate and train classifier
clf = RandomForestClassifier(random_state=0)
clf.fit(X, y)

# print class predictions of the training data
print(clf.predict(X))
```

```python
test1 = [[r,0.6,0.3] for r in np.arange(0.0, 1.0, 0.05)]
test2 = [[0.4,g,0.3] for g in np.arange(0.0, 1.0, 0.05)]
print(clf.predict(test2))
```
